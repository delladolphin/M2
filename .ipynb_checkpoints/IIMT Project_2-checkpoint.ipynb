## (A) Understanding the data & Data Wrangling

```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet, RidgeCV, LassoCV, ElasticNetCV
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from scipy.stats import norm
import statsmodels.api as sm
from scipy import stats
import datetime
import random

np.random.seed(42)
random.seed(42)

df = pd.read_csv('detailed listing.csv')
```

```python
df.shape
```

```python
df.info()
```

```python
df["reviews_per_month"].fillna(0 ,inplace = True)
df["last_review"].fillna("0" ,inplace = True)
df.info()
```

```python
df = df.apply(lambda x:x.fillna(x.value_counts().index[0]))
df.head()
```

```python
df.values
```

```python
df.describe()
```

```python
df = df.drop(['id', 'host_id','host_name'], axis=1)
df.head()
```

```python
df.dtypes
```

```python
#Change types and normalize the availability.
df['price'] = df['price'].astype(float)
df['minimum_nights'] = df['minimum_nights'].astype(float)
df['number_of_reviews'] = df['number_of_reviews'].astype(float)
df['reviews_per_month'] = df['reviews_per_month'].astype(float)
df['calculated_host_listings_count'] = df['calculated_host_listings_count'].astype(float)
df['availability_365'] = df['availability_365'].astype(float)
df['availability_365'] = df['availability_365'] / 365

df["last_review"] = pd.to_datetime(df["last_review"])
df["last_review"].dt
df["last_review"].dt.year
df["last_review"].dt.month
df["last_review"].dt.minute

df.dtypes
```

```python
Y= Train["price"]
X= Train['latitude'],['longitude'],['minimum_nights'],['number_of_reviews'],['calculated_host_listings_count'],['availability_365']


from sklearn import linear_model
# with sklearn
regr = linear_model.LinearRegression()
regr.fit(X, Y)

print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)

```

```python
import matplotlib as mpl 
from mpl_toolkits.mplot3d import Axes3D 
import matplotlib.pyplot as plt 
```

```python
def generate_dataset(n): 
    y= Train["price"]
    x= Train['latitude'],['longitude'],['minimum_nights'],['number_of_reviews'],['calculated_host_listings_count'],['availability_365']

    random_x1 = np.random.rand() 
    random_x2 = np.random.rand() 
    for i in range(n): 
        x1 = i 
        x2 = i/2 + np.random.rand()*n 
        x.npappend([1, x1, x2]) 
        y.npappend(random_x1 * x1 + random_x2 * x2 + 1) 
    return np.array(x), np.array(y) 
  
x, y = generate_dataset(200) 
  
mpl.rcParams['legend.fontsize'] = 12
  
fig = plt.figure() 
ax = fig.gca(projection ='3d') 
  
ax.scatter(x[:, 1], x[:, 2], y, label ='y', s = 5) 
ax.legend() 
ax.view_init(45, 0) 
  
plt.show() 
```
